**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
camden_2018 = dfg[(dfg['London_Borough']=='Camden') & (dfg['Year'] == 2018)]
camden_2018
```

```python
camden_1998 = dfg[(dfg['London_Borough']=='Camden') & (dfg['Year'] == 1998)]
camden_1998
```

```python
dfg.loc[dfg['London_Borough'] == 'Camden']
```

```python
df['London_Borough'].unique()
```

```python
def create_price_ratio(d):
    y1998 = float(d.loc[d['Year'] == 1998, 'Average_price'].iloc[0])
    y2018 = float(d.loc[d['Year'] == 2018, 'Average_price'].iloc[0])
    ratio = y2018 / y1998
    return ratio 
```

```python
final = {}
```

```python
for borough_name in df['London_Borough'].unique():
    borough_data = dfg[dfg['London_Borough'] == borough_name]
    ratio = create_price_ratio(borough_data)
    final[borough_name] = ratio
```

```python
df_ratios = pd.DataFrame(list(final.items()), columns=['London_Borough', 'Price_Ratio'])
df_ratios.sort_values(by='Price_Ratio', ascending=False).head(34).plot(
    kind='bar', x='London_Borough', y='Price_Ratio', title='House Price Growth (2018 vs 1998)', legend=False
)
```

```python
#1- add 1998 and 2018 columns side by side
compare = (
    dfg[dfg['Year'].isin([1998, 2018])]
      .pivot(index='London_Borough', columns='Year', values='Average_price')
)

# focus on top 'n' boroughs (or by growth)
compare = compare.sort_values(2018, ascending=False)

#2- lot recent times (2018 in the back and 1998 in the front)
x = np.arange(len(compare))
w_back  = 0.8   # width of 2018 bars
w_front = 0.55  # width of 1998 bars (in front)

fig, ax = plt.subplots(figsize=(14, 7), dpi=120)

# back layer (2018)
ax.bar(x, compare[2018].values, width=w_back, label='2018', alpha=0.8)

#orange bars in font. 
ax.bar(x, compare[1998].values, width=w_front, label='1998', alpha=0.95)

ax.set_xticks(x)
ax.set_xticklabels(compare.index, rotation=60, ha='right')
ax.set_ylabel('Average price (£ millions)')
ax.set_title('London Borough House Prices: 1998 vs 2018 (overlay)')
ax.legend()
ax.margins(x=0.01)
fig.tight_layout()
plt.show()
```
